```python
#@title Funktionen für Spaltenerstellung

# Liste der zu checkenden Spalten
echt_columns = [f'Echt{i}' for i in range(1, 31)]
fake_columns = [f'Fake{i}' for i in range(1, 31)]

# Funktion um korrekte Antworten kenntlich zu machen
def mark_correct_answers(row):
    new_row = {}
    for col in echt_columns:
        if col in row:
            new_row[f'{col}_correct'] = row[col] == 'echt'
    for col in fake_columns:
        if col in row:
            new_row[f'{col}_correct'] = row[col] == 'künstlich generiert'
    return pd.Series(new_row)


# Funktion zum berechnen der False Werte
def calculate_false_percentages(row):
    correct_columns = [col for col in df_cleaned.columns if col.endswith('_correct')]
    echt_correct_columns = [col for col in correct_columns if col.startswith('Echt')]
    fake_correct_columns = [col for col in correct_columns if col.startswith('Fake')]

    # Berechne overall Fehlerquoten
    total = len(correct_columns)
    false_count = sum(row[col] == False for col in correct_columns)
    overall_false_percentage = (false_count / total) * 100 if total > 0 else 0

    # Berechne Fehlerquoten nur für Echte Bilder
    total_echt = len(echt_correct_columns)
    false_count_echt = sum(row[col] == False for col in echt_correct_columns)
    echt_false_percentage = (false_count_echt / 30) * 100 if total > 0 else 0

    # Berechne Fehlerquoten für KI Bilder
    total_fake = len(fake_correct_columns)
    false_count_fake = sum(row[col] == False for col in fake_correct_columns)
    fake_false_percentage = (false_count_fake / 30) * 100 if total > 0 else 0

    return pd.Series({
        'False_percentage': overall_false_percentage,
        'Echt_false_percentage': echt_false_percentage,
        'Fake_false_percentage': fake_false_percentage
    })


# Anpassung der Spalten für Exel
def adjust_column_widths(filename):
    workbook = load_workbook(filename)
    sheet = workbook.active
    for column in sheet.columns:
        max_length = 0
        column_letter = get_column_letter(column[0].column)
        for cell in column:
            try:
                if len(str(cell.value)) > max_length:
                    max_length = len(cell.value)
            except:
                pass
        adjusted_width = (max_length + 2)
        sheet.column_dimensions[column_letter].width = adjusted_width
    workbook.save(filename)

# Funktion zum berechnen der falsch klassifizierten Echten Bilder
def count_false_echt(row):
    echt_correct_columns = [col for col in df_cleaned.columns if col.startswith('Echt') and col.endswith('_correct')]
    return sum(row[col] == False for col in echt_correct_columns)

# Funktion zum berechnen der falsch klassifizierten KI Bilder
def count_false_fake(row):
    fake_correct_columns = [col for col in df_cleaned.columns if col.startswith('Fake') and col.endswith('_correct')]
    return sum(row[col] == False for col in fake_correct_columns)

# Funktion für Altersgruppen Spalte
def add_age_group_column(df, age_column='Alter'):
    conditions = [
        (df[age_column] == '50-59 Jahre alt') | (df[age_column] == '60-69 Jahre alt')
    ]
    choices = ['old']

    df['Age_Group'] = np.select(conditions, choices, default='young')

    return df

# Funktion zum Mapping von Zuversicht zu Integern
def map_confidence_to_int(confidence_level):
    mapping = {
        'sehr unsicher': 1,
        'unsicher': 2,
        'neutral': 3,
        'sicher': 4,
        'sehr sicher': 5
    }
    return mapping.get(confidence_level, 0)

# Funktion um das Mapping anzuwenden
def add_confidence_int_columns(df):
    df['ConEcht_Int'] = df['ConEcht'].apply(map_confidence_to_int)
    df['ConFake_Int'] = df['ConFake'].apply(map_confidence_to_int)

    return df


# Liste der EchtTime und FakeTime Spalten
echt_time_columns = [col for col in df_cleaned.columns if col.startswith('Echt') and col.endswith('Time')]
fake_time_columns = [col for col in df_cleaned.columns if col.startswith('Fake') and col.endswith('Time')]
all_time_columns = ['AlterTime', 'ValidTime'] + echt_time_columns + fake_time_columns
```

```python
#@title Anwendung von Dataframe Operationen

# Anwendung von Spaltenmarkierung als echt oder falsch
correct_columns = df_cleaned.apply(mark_correct_answers, axis=1)
df_cleaned = pd.concat([df_cleaned, correct_columns], axis=1)

# Funktion zum Fehlerquoten Berechnung anwenden
false_percentages = df_cleaned.apply(calculate_false_percentages, axis=1)
df_cleaned = pd.concat([df_cleaned, false_percentages], axis=1)

# Neue Spalten für False counts
df_cleaned['False_count_echt'] = df_cleaned.apply(count_false_echt, axis=1)
df_cleaned['False_count_fake'] = df_cleaned.apply(count_false_fake, axis=1)

# Berechne average Zeiten
df_cleaned['Overall_avg_time'] = df_cleaned[all_time_columns].mean(axis=1)
df_cleaned['Echt_avg_time'] = df_cleaned[echt_time_columns].mean(axis=1)
df_cleaned['Fake_avg_time'] = df_cleaned[fake_time_columns].mean(axis=1)

# Füge die Altersgruppen Spalte hinzu
df_cleaned = add_age_group_column(df_cleaned)

# Zuversicht als Integer als Spalte hinzufügen
df_cleaned = add_confidence_int_columns(df_cleaned)
```

```python
#@title One sample t-Test Genauigkeitsraten

# Daten definieren
observed_accuracy = 0.8701  # 87.01% in Dezimalform
n = 112  # Anzahl der Teilnehmer nach Datenbereinigung
null_hypothesis = 0.5  # 50% Zufallsrate

# Standardfehler berechnen
# Wir nehmen an, dass jede Beobachtung entweder korrekt (1) oder inkorrekt (0) ist
standard_error = np.sqrt((observed_accuracy * (1 - observed_accuracy)) / n)

# t-Statistik manuell berechnen
t_statistic = (observed_accuracy - null_hypothesis) / standard_error

# p-Wert berechnen
freedom = n - 1  # Freiheitsgrade
p_value = stats.t.sf(abs(t_statistic), freedom) * 2  # Zweiseitiger Test

# Ergebnisse ausgeben
print(f"Beobachtete Genauigkeit: {observed_accuracy:.4f}")
print(f"Anzahl der Teilnehmer: {n}")
print(f"t-Statistik: {t_statistic:.4f}")
print(f"p-Wert: {p_value:.4e}")
print(f"Freiheitsgrade (df): {freedom}")
```

```python
#@title Alle Two Way Anova Tabellen

# Liste der abhängigen Variablen
dependent_vars = ['False_percentage', 'Echt_false_percentage', 'Fake_false_percentage',
                  'Overall_avg_time', 'Echt_avg_time', 'Fake_avg_time']

# DataFrame zum speichern der Ergebnisse
anova_results = pd.DataFrame()

# Two-Way-Mixes-ANOVA für alle abhängigen Variablen
for dv in dependent_vars:
    formula = f'{dv} ~ C(Age_Group) + C(Geschlecht) + C(Age_Group):C(Geschlecht)'
    model = ols(formula, data=df_cleaned).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # abhängige Variablen hinzufügen
    anova_table['Dependent Variable'] = dv

    # Ergebnisse anhängen
    anova_results = pd.concat([anova_results, anova_table.reset_index()], ignore_index=True)

# Spalten umordnen
anova_results = anova_results[['Dependent Variable', 'index', 'sum_sq', 'df', 'F', 'PR(>F)']]
anova_results.columns = ['Dependent Variable', 'Source', 'Sum of Squares', 'Degrees of Freedom', 'F Value', 'p-value']

# Plot vorbereiten
plt.figure(figsize=(12, 6))
plt.axis('tight')
plt.axis('off')
sns.set(style="whitegrid")

# Eine Tabelle erstellen
table = plt.table(cellText=anova_results.values, colLabels=anova_results.columns, cellLoc='center', loc='center')

# Style der Tabelle
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)

# Zeige den Plot
plt.show()
```

```python
#@title Two Way Mixed Anova für Alter und Echtheit im Bezug auf Genauigkeit
# NOTE: TWO-WAY ANALYSIS, ACCURACY FOR CHIRA24

dependent_vars_error = ['Echt_false_percentage',
'Fake_false_percentage','Age_Group']

filtered_df_error = df_cleaned[dependent_vars_error]

# avoid SettingWithCopyWarning:
filtered_df_error = filtered_df_error.copy()

# userId hinzufügen
filtered_df_error['userID'] = list(range(1, len(filtered_df_error) + 1))

# melt zu long format:
df_long_error = pd.melt(filtered_df_error, id_vars=['Age_Group','userID'], value_vars=['Echt_false_percentage', 'Fake_false_percentage'],
                  var_name='Echtheit', value_name='Percentage')

# quick check
print(df_long_error.loc[df_long_error['userID'] == 1])
print(df_long_error.loc[df_long_error['userID'] == 33])
print(df_long_error.loc[df_long_error['userID'] == 42])
print('\n')

result_sphericity=pg.sphericity(data=df_long_error, dv='Percentage', within='Echtheit', subject='userID', method='mauchly', alpha=0.05)

print(result_sphericity)

# Führen Sie Ihre ANOVA-Analyse durch
result_error = pg.mixed_anova(dv='Percentage', between='Age_Group',
                              within='Echtheit', subject='userID',
                              correction='auto', data=df_long_error)

print(result_error)

# Runden Sie die Ergebnisse
result_error_rounded = result_error.round(3)

# Erstellen Sie ein neues DataFrame mit den umbenannten Indizes
new_data = {
    'Source': ['Altersgruppe', 'Echtheit', 'Interaktion'],
    'SS': result_error_rounded['SS'],
    'DF1': result_error_rounded['DF1'],
    'DF2': result_error_rounded['DF2'],
    'MS': result_error_rounded['MS'],
    'F': result_error_rounded['F'],
    'p-unc': result_error_rounded['p-unc'],
    'np2': result_error_rounded['np2'],
    'eps': result_error_rounded['eps']
}

new_result = pd.DataFrame(new_data)

# Erstellen Sie die Tabelle
fig, ax = plt.subplots(figsize=(12, 2))
ax.axis('off')

tbl = ax.table(cellText=new_result.values,
               colLabels=new_result.columns,
               cellLoc='center', loc='center')

plt.title('Two-Way Mixed ANOVA für Alter und Echtheit auf die Genauigkeit', fontsize=14, pad=-10)

tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.scale(1.2, 1.5)

plt.tight_layout()
plt.show()
```

```python
# Zuversichtsspalten auswählen
dependent_vars_confidence = ['ConEcht_Int', 'ConFake_Int']

filtered_df_confidence = df_cleaned[dependent_vars_confidence]

# avoid SettingWithCopyWarning:
filtered_df_confidence = filtered_df_confidence.copy()

# userId hinzufügen
filtered_df_confidence['userID'] = list(range(1, len(filtered_df_confidence) + 1))

# melt zu long format:
df_long_confidence = pd.melt(filtered_df_confidence, id_vars=['userID'],
                             value_vars=['ConEcht_Int', 'ConFake_Int'],
                             var_name='Echtheit', value_name='Zuversicht')

# quick check
print(df_long_confidence.loc[df_long_confidence['userID'] == 1])
print(df_long_confidence.loc[df_long_confidence['userID'] == 33])
print(df_long_confidence.loc[df_long_confidence['userID'] == 42])
print('\n')

# Mann-Whitney U-Test für Zuversicht
real_confidence = df_long_confidence[df_long_confidence['Echtheit'] == 'ConEcht_Int']['Zuversicht']
fake_confidence = df_long_confidence[df_long_confidence['Echtheit'] == 'ConFake_Int']['Zuversicht']
u_stat, p_value = stats.mannwhitneyu(real_confidence, fake_confidence)

print("Mann-Whitney U Test für Zuversicht:")
print(f"U-Statistik: {u_stat}")
print(f"p-Wert: {p_value}")

# Darstellung der Ergebnisse
results = {
    'Analyse': ['Echtheitseffekt'],
    'U-Statistik': [u_stat],
    'p-Wert': [p_value]
}

results_df = pd.DataFrame(results)

fig, ax = plt.subplots(figsize=(12, 1))
ax.axis('off')

tbl = ax.table(cellText=results_df.values,
               colLabels=results_df.columns,
               cellLoc='center', loc='center')

plt.title('Mann-Whitney U Test für Zuversicht nach Echtheit der Bilder', fontsize=14, pad=-10)

tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.scale(1.2, 1.5)

plt.tight_layout()
plt.show()
```

```python
#@title Mann Whitney U Tabelle

# Altersgruppen definieren
age_group_old = df_cleaned[df_cleaned['Age_Group'] == 'old']
age_group_young = df_cleaned[df_cleaned['Age_Group'] == 'young']

# Mann-Whitney U Test für ConEcht_Int (Zuversicht in echte Bilder)
u_statistic_echt, p_value_echt = mannwhitneyu(age_group_old['ConEcht_Int'], age_group_young['ConEcht_Int'])

# Mann-Whitney U Test für ConFake_Int (Zuversicht in KI-Bilder)
u_statistic_fake, p_value_fake = mannwhitneyu(age_group_old['ConFake_Int'], age_group_young['ConFake_Int'])

print(p_value_echt*3)
print(p_value_fake*3)

# round Values
u_statistic_echt = round(u_statistic_echt, 3)
u_statistic_fake = round(u_statistic_fake, 3)
p_value_echt = round(p_value_echt, 3)
p_value_fake = round(p_value_fake, 3)

# Bonferroni-Korrektur durchführen
num_tests = 3
bonferroni_alpha = 0.05 * num_tests

# Bonferroni-korrigierte p-Werte berechnen
corrected_p_value_echt = round(min(p_value_echt * num_tests, 1.0), 3)
corrected_p_value_fake = round(min(p_value_fake * num_tests, 1.0), 3)

# Ergebnisse in einem DataFrame darstellen
mann_whitney_results = {
    'Analyse': ['Zuversicht in echte Bilder', 'Zuversicht in KI-Bilder'],
    'U-Statistik': [u_statistic_echt, u_statistic_fake],
    'p-Wert': [p_value_echt, p_value_fake],
    'Bonferroni-korrigierter p-Wert': [corrected_p_value_echt, corrected_p_value_fake]
}

mann_whitney_df = pd.DataFrame(mann_whitney_results)

# Tabelle plotten und anzeigen
fig, ax = plt.subplots(figsize=(12, 1))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=mann_whitney_df.values, colLabels=mann_whitney_df.columns, cellLoc='center', loc='center')

# Überschrift für Mann-Whitney U-Test-Tabelle als Text hinzufügen
plt.title('Mann-Whitney U Test für Effekt des Alters auf Zuversicht', pad=0)

# Tabelle anzeigen
plt.show()
```

```python
#@title Funktion zur Berechnung der individuellen Bildleistungen

def calculate_picture_performance_by_age(df):
    results = {}

    # Gruppieren nach Altersgruppe und Iteration durch jede Gruppe
    for age_group, group_df in df.groupby('Age_Group'):
        group_results = {}

        # Iteration durch Spalten und Berechnung der Leistung für jedes Bild innerhalb der Altersgruppe
        for col in group_df.columns:
            if '_correct' in col:  # Nur 'correct' Spalten verarbeiten
                # Konvertiere die Spalte in numerisch, wobei nicht-numerische Werte zu NaN werden
                group_df[col] = pd.to_numeric(group_df[col], errors='coerce')

                true_count = group_df[col].sum()
                false_count = (group_df[col] == 0).sum()  # Zähle explizite 0-en als Falsch
                total_count = group_df[col].notna().sum()  # Zähle nicht-NaN Werte für die Gesamtzahl

                true_percentage = (true_count / total_count) * 100 if total_count > 0 else 0
                false_percentage = (false_count / total_count) * 100 if total_count > 0 else 0

                # Speichere Ergebnisse für dieses Bild innerhalb der Altersgruppe
                group_results[col + '_True_Percentage'] = true_percentage
                group_results[col + '_False_Percentage'] = false_percentage

        # Speichere die Ergebnisse der Gruppe im Hauptergebnisverzeichnis
        results[age_group] = group_results

    return results

# Anwendung
results_by_age = calculate_picture_performance_by_age(df_cleaned.copy())

# Drucke die Ergebnisse
for age_group, group_results in results_by_age.items():
    print(f"Ergebnisse für Altersgruppe: {age_group}")
    for key, value in group_results.items():
        print(f"  {key}: {value:.2f}%")
    print()
```

```python
#@title Lineplot kombiniert

def plot_picture_performance_combined_sorted(results_by_age):
    sns.set_theme(style="whitegrid")
    colors = sns.color_palette("colorblind")

    # Daten vorbereiten
    data = []
    for age_group, group_results in results_by_age.items():
        for pic_type in ['Echt', 'Fake']:
            for pic, value in group_results.items():
                if pic_type in pic and 'False_Percentage' in pic:
                    data.append({
                        'Age_Group': age_group,
                        'Picture_Type': pic_type,
                        'Picture_Number': pic.split('_')[0],
                        'Error_Rate': value
                    })

    df = pd.DataFrame(data)

    # Plot erstellen
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

    # Farbpalette initialisieren
    age_palette = {'young': colors[0], 'old': colors[1]}

    # Title
    titles = {
        'Echt': 'Fehlerquote (%) der echten Bilder',
        'Fake': 'Fehlerquote (%) der KI generierten Bilder'
    }

    for pic_type, ax in zip(['Echt', 'Fake'], [ax1, ax2]):
        # Filtern und sortieren der Daten
        df_pic = df[df['Picture_Type'] == pic_type].copy()
        df_pic['Avg_Error_Rate'] = df_pic.groupby('Picture_Number')['Error_Rate'].transform('mean')
        df_pic = df_pic.sort_values('Avg_Error_Rate')

        # Plot
        sns.lineplot(data=df_pic, x='Picture_Number', y='Error_Rate',
                     hue='Age_Group', marker='o', ax=ax, palette=age_palette)
        ax.set_title(titles[pic_type])
        ax.set_xlabel('Bildnummer (sortiert nach durchschnittlicher Fehlerquote)')
        ax.set_ylabel('Fehlerquote (%)')
        ax.set_ylim([0, 100])
        ax.tick_params(axis='x', rotation=90, labelsize='small')
        ax.set_yticks(range(0, 101, 5))

    plt.tight_layout()
    plt.show()

# Rufe die Plotting-Funktion mit dem results_by_age Dictionary auf
plot_picture_performance_combined_sorted(results_by_age)
```

```python
# Funktion zur Darstellung der durchschnittlichen Antwortzeiten nach Altersgruppen und Bildtypen
def plot_avg_time_by_age(df):
    sns.set_theme(style="whitegrid")
    colors = sns.color_palette("colorblind")

    # Bestimme die Y-Achsen-Grenzen
    yticks = range(0, 16, 1)  # Passe die Schrittgröße bei Bedarf an

    # Daten umstrukturieren für die Verwendung von hue in einem einzigen Plot
    df_melted = df.melt(id_vars=['Age_Group'],
                        value_vars=['Echt_avg_time', 'Fake_avg_time'],
                        var_name='Bildtyp',
                        value_name='Durchschnittliche Zeit')

    # Bezeichnungen ändern
    df_melted['Bildtyp'] = df_melted['Bildtyp'].replace({'Echt_avg_time': 'Echte Bilder', 'Fake_avg_time': 'KI Bilder'})

    # Erstelle ein einzelnes Diagramm mit Schattierung
    plt.figure(figsize=(14, 6))
    sns.lineplot(data=df_melted, x='Age_Group', y='Durchschnittliche Zeit', hue='Bildtyp', palette=colors, marker='o', ci=95)
    plt.title('Durchschnittliche Zeit zum Beantworten der Bilder nach Altersgruppe und Bildtyp')
    plt.xlabel('Altersgruppen')
    plt.ylabel('Durchschnittliche Zeit (Sekunden)')
    plt.ylim([0, 16])
    plt.yticks(yticks)
    plt.xticks(rotation=45)
    plt.legend(title='Bildtyp')
    plt.tight_layout()
    plt.show()

#Rufe den Plot auf
plot_avg_time_by_age(df_cleaned)
```

```python
#@title Zeiten nach Alter und Geschlecht

def plot_avg_time_by_age(df):
    sns.set_theme(style="whitegrid")
    colors = sns.color_palette("colorblind")

    # y-axis limits
    yticks = range(0, 16, 1)

    # Sublot für jeden Bildertypen (Echt and Fake)
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Plot für Echt Bilder average time
    sns.lineplot(data=df, x='Age_Group', y='Echt_avg_time', ax=axes[0], hue='Geschlecht', palette=colors, marker='o', ci=None)
    axes[0].set_title('Durchschnittliche Zeit zum beantworten der echten Bilder nach Altergruppe')
    axes[0].set_xlabel('Altergruppen')
    axes[0].set_ylabel('Durchschnittliche Zeit (Sekunden)')
    axes[0].tick_params(axis='x', rotation=45)
    axes[0].legend(title='Geschlecht')
    axes[0].set_ylim([0, 16])
    axes[0].set_yticks(yticks)

    # Plot für Fake Bilder average time
    sns.lineplot(data=df, x='Age_Group', y='Fake_avg_time', ax=axes[1], hue='Geschlecht', palette=colors, marker='o', ci=None)
    axes[1].set_title('Durchschnittliche Zeit zum beantworten der KI Bilder nach Altergruppe')
    axes[1].set_xlabel('Altergruppen')
    axes[1].set_ylabel('Durchschnittliche Zeit (Sekunden)')
    axes[1].tick_params(axis='x', rotation=45)
    axes[1].legend(title='Geschlecht')
    axes[1].set_ylim([0, 16])
    axes[1].set_yticks(yticks)

    plt.tight_layout()
    plt.show()

plot_avg_time_by_age(df_cleaned)
```
